# 01 Data Ingestion
Load, inspect, and validate all 10 mutual fund datasets.

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
DATA_PATH = Path('../data/raw')
```

```python
datasets = {
    "fund_master":"fund_master.csv",
    "nav_history": "nav_history.csv",
    "returns_summary": "returns_summary.csv",
    "portfolio_holdings": "portfolio_holdings.csv",
    "benchmark_nifty100": "benchmark_nifty100.csv",
    "sip_transactions": "sip_transactions.csv",
    "lumpsum_transactions": "lumpsum_transactions.csv",
    "investor_profiles": "investor_profiles.csv",
    "dividend_history": "dividend_history.csv",
    "risk_metrics": "risk_metrics.csv"
}

dataframes = {}

for name, file in datasets.items():
    dataframes[name] = pd.read_csv(DATA_PATH / file)
    print(f"\n{name}")
    print("Shape:", dataframes[name].shape)
    display(dataframes[name].head())
```

## Data Types Inspection

```python
for name, df in dataframes.items():
    print(f"\n{name}")
    print(df.dtypes)
```

## Missing Values Check

```python
for name, df in dataframes.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

## Duplicate Records Check

```python
for name, df in dataframes.items():
    print(f"{name}: {df.duplicated().sum()} duplicates")
```

## Fund Master Exploration

```python
fund_master = dataframes['fund_master']

print('Fund Houses')
print(fund_master['fund_house'].unique())

print('\nCategories')
print(fund_master['category'].unique())

print('\nSub Categories')
print(fund_master['sub_category'].unique())

print('\nRisk Grades')
print(fund_master['risk_grade'].unique())
```

## AMFI Code Validation

```python
master_codes = set(fund_master['amfi_code'])
nav_codes = set(dataframes['nav_history']['amfi_code'])

missing_codes = master_codes - nav_codes

print('Missing Codes:', len(missing_codes))
print(missing_codes)
```

## Data Quality Summary

```python
quality_summary = pd.DataFrame({
    'Metric':[
        'Total Schemes',
        'Duplicate Funds',
        'Missing Fund Values',
        'Total NAV Records',
        'Duplicate NAV Records',
        'Missing NAV Values',
        'Missing AMFI Codes'
    ],
    'Value':[
        len(fund_master),
        fund_master.duplicated().sum(),
        fund_master.isnull().sum().sum(),
        len(dataframes['nav_history']),
        dataframes['nav_history'].duplicated().sum(),
        dataframes['nav_history'].isnull().sum().sum(),
        len(missing_codes)
    ]
})

quality_summary
```
